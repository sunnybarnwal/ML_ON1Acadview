```
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mpl
```

```
dta=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
```

```
dta.head()
```

```
sb.countplot(x='Temp3pm',data=dta)
```

```
sb.countplot(x='Rainfall',data=dta)
```

```
sb.boxplot(x='Cloud9am',y='Cloud3pm',data=dta)
```

```
dta["Rainfall"].plot(kind="barh")
```

```
dta["WindSpeed9am"].plot(kind="bar")
```

```
sb.countplot(x="RainToday",data=dta)
mpl.title("Will there be rain today or not")
```

```
sb.distplot(dta["MaxTemp"])
```

```
sb.regplot(x='Humidity9am', y='Pressure9am', data=dta)
```

```
dta["WindSpeed9am"].value_counts().plot(kind="pie")
```

```
dta['Rainfall'].plot(kind='hist')
```

```
sb.distplot(dta['Sunshine'].dropna())
```

```
dta['Evaporation'].plot(kind='hist')
```

```
sb.violinplot(x='Rainfall',data=dta)
```

```
sb.violinplot(x='Rainfall',y='Evaporation',data=dta)
mpl.title('Evaporation vs rainfall violinplot')
```

```
sb.violinplot(x='RainToday',y='MaxTemp',data=dta)
mpl.title('Raintoday vs Maxtemp violinplot')
```

```
sb.jointplot(x="WindSpeed9am",y="Humidity9am",data=dta)
```

```
sb.jointplot(x="WindSpeed9am",y="Humidity9am",data=dta,kind='hex')
```
